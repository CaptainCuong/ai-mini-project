## Word2Vec

## Import libs

```python
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
% matplotlib inline
import nltk
nltk.download('punkt')
from gensim.models import KeyedVectors

################## Preprocessing #######################
import re
import codecs
import multiprocessing
import gensim
from gensim.models import Word2Vec

# Comment this out if you want to see Deprecation warnings
import warnings
warnings.simplefilter("ignore", DeprecationWarning)
```

```python
def preprocess_text(text):
    text = re.sub('[^a-zA-Zа-яА-Я1-9]+', ' ', text)
    text = re.sub(' +', ' ', text)
    return text.strip()


def prepare_for_w2v(filename_from, filename_to, lang):
    raw_text = codecs.open(filename_from, "r", encoding='utf-8').read()
    with open(filename_to, 'w', encoding='utf-8') as f:
        for sentence in nltk.sent_tokenize(raw_text, lang):
            print(preprocess_text(sentence.lower()), file=f)
            

def train_word2vec(filename):
    data = gensim.models.word2vec.LineSentence(filename)
    return Word2Vec(data, size=300, window=5, min_count=3, workers=multiprocessing.cpu_count())
```

## Train and save word2vec model into GG drive on Game of Thrones Dataset

```python
prepare_for_w2v('gameofthrones.txt',  'train_got.txt', 'english')
model= train_word2vec('train_got.txt')
model.save('model_got.kv')
```

## Load word2vec model from GG drive

```python
model_ak = KeyedVectors.load('model_got.kv')
```

## Visualizing 300-dimensonal word vector onto 2-dimensional using dimension reduction

```python
def plot_similar_words(title, labels, embedding_clusters, word_clusters, a, filename=None):
    plt.figure(figsize=(16, 9))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embeddings, words, color in zip(labels, embedding_clusters, word_clusters, colors):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, c=color.reshape(1,-1), alpha=a, label=label)
        for i, word in enumerate(words):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                         textcoords='offset points', ha='right', va='bottom', size=8)
    plt.legend(loc=4)
    plt.title(title)
    plt.grid(True)
    if filename:
        plt.savefig(f'images/{filename}', format='png', dpi=150, bbox_inches='tight')
    plt.show()
```

```python
keys = ['game', 'war', 'good', 'bad', 'king', 'queen', 'expensive',
        'experience', 'food', 'peace', 'release', 'strong']
target = []
embedding_clusters = []
word_clusters = []
for word in keys:
    embeddings = []
    words = []
    for similar_word, _ in model_ak.wv.most_similar(word, topn=30):
        words.append(similar_word)
        embeddings.append(model_ak.wv[similar_word])
        target.append(word)
        # embeddings.append(model_gn[similar_word])
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
```

```python
embedding_clusters = np.array(embedding_clusters)
n, m, k = embedding_clusters.shape
tsne_ak_2d = TSNE(perplexity=10, n_components=2, init='pca', n_iter=3500, random_state=32, early_exaggeration=20)
embeddings_ak_2d_keys =  np.array(tsne_ak_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
```

### TSNE

```python
plot_similar_words('t-SNE', keys, embeddings_ak_2d_keys, word_clusters, 0.7,
                        'similar_words_tsne.png')
```

### PCA

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
pca = PCA(n_components=2)
X = np.array(embedding_clusters).reshape(n*m,k)
X_r = pca.fit(X).transform(X).reshape(n, m, 2)

plot_similar_words('PCA', keys, X_r, word_clusters, 0.7,
                        'similar_words_PCA.png')
```

### LDA

```python
y = target
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X).reshape(n, m, 2)
plot_similar_words('LDA', keys, X_r2, word_clusters, 0.7,
                        'similar_words_LDA.png')
```

## Evaluate model

```python
perplexities = [150, 200, 250, 300]

# fig, ax = plt.subplots(2, 2, figsize = 20,4)
for perplexity in perplexities:
  embedding_clusters = np.array(embedding_clusters)
  n, m, k = embedding_clusters.shape
  tsne_ak_2d = TSNE(perplexity=perplexity, n_components=2, init='pca', n_iter=3500, random_state=32)
  embeddings_ak_2d_keys =  np.array(tsne_ak_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
  plot_similar_words(f't-SNE, perplexity={perplexity}', keys, embeddings_ak_2d_keys, word_clusters, 0.7,
                        f'images/similar_words_tsne_per{perplexity}.png')

  
```

## GoogleNews-vectors-negative30 Dataset

```python
!wget -c "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"

!gzip -d GoogleNews-vectors-negative300.bin.gz
import gensim

model_gn = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
keys = ['Paris', 'Python', 'Sunday', 'Tolstoy', 'Twitter', 'bachelor', 'delivery', 'election', 'expensive',
        'experience', 'financial', 'food', 'iOS', 'peace', 'release', 'war']
embedding_clusters = []
word_clusters = []
for word in keys:
    embeddings = []
    words = []
    for similar_word, _ in model_gn.most_similar(word, topn=30):
        words.append(similar_word)
        embeddings.append(model_gn[similar_word])
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
```

```python
embedding_clusters = np.array(embedding_clusters)
n, m, k = embedding_clusters.shape
tsne_model_en_2d = TSNE(perplexity=15, n_components=2, init='pca', n_iter=1000, random_state=32)
embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
```

```python
plot_similar_words('Similar words from Google News', keys, embeddings_en_2d, word_clusters, 0.7,
                        'similar_words.png')
```
